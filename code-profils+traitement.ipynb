```python
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import WebDriverException

# Initialize lists to store extracted data
names = []
companies = []
locs = []
certifs = []
final_formations = []

for link in L:
    try:
        # Load the profile page
        driver.get(link)
        time.sleep(8)
        src = driver.page_source
        soup = BeautifulSoup(src, 'lxml')

        # Retrieve the name
        name_element = soup.find('h1', {'class': 'text-heading-xlarge inline t-24 v-align-middle break-words'})
        if name_element is not None:
            name = name_element.get_text().strip()
        else:
            name = ""
            print("Name element not found on the page.")

        # Retrieve the location
        loc_element = soup.find('span', {'class': 'text-body-small inline t-black--light break-words'})
        if loc_element is not None:
            loc = loc_element.get_text().strip()
        else:
            loc = ""
            print("Location element not found on the page.")

        # Retrieve the company
        company_element = soup.find('div', {'class': 'inline-show-more-text'})
        if company_element is not None:
            company = company_element.get_text().strip()
        else:
            company = ""
            print("Company element not found on the page.")

        try:
            full_link_certifications = link + "details/certifications/"
            driver.get(full_link_certifications)
            time.sleep(8)
            src_certifications = driver.page_source
            soup_certifications = BeautifulSoup(src_certifications, 'lxml')

            # Retrieve the certifications
            certifs_divs = soup_certifications.find_all('div', class_='display-flex flex-row justify-space-between')
            certifications = []
            for div in certifs_divs:
                competence_span = div.find('span', class_='mr1 hoverable-link-text t-bold')
                if competence_span:
                    competence = competence_span.find('span', attrs={"aria-hidden": "true"})
                    if competence:
                        certifications.append(competence.text.strip())

            # Convert the list of certifications into a comma-separated string
            certifications_str = ", ".join(certifications)

        except WebDriverException as e:
            certifications_str = ""
            print(f"Error accessing certifications: {e}")

        # Construct the full education link
        full_link_education = link + "details/education/"

        # Load the education page
        driver.get(full_link_education)
        time.sleep(8)
        src_education = driver.page_source
        soup_education = BeautifulSoup(src_education, 'html.parser')

        # Retrieve the extracted formations
        extracted_formations = []
        elements_education = soup_education.find_all('span', class_='mr1 hoverable-link-text t-bold')
        for element in elements_education:
            formation = element.text.strip()
            extracted_formations.append(formation)

        # Check the first position of the extracted formations list
        if extracted_formations and extracted_formations[0] == "Ecole Nationale Sup√©rieure d'Informatique et d'Analyse des Syst√®mes - ENSIASEcole Nationale Sup√©rieure d'Informatique et d'Analyse des Syst√®mes - ENSIAS":
            result = "No"
        else:
            result = "DD"

        # Append the extracted data to the lists
        names.append(name)
        companies.append(company)
        locs.append(loc)
        certifs.append(certifications_str)
        final_formations.append(result)

    except WebDriverException as e:
        print(f"Error accessing profile: {e}")
```

```python
# Create a DataFrame from the extracted data
data = {
    "Name": names,
    "Company": companies,
    "Location": locs,
    "Certifications": certifs,
    "Formations": final_formations
}
df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
# Save the DataFrame to a CSV file
df.to_csv("linkedin_data.csv", index=False)
```
